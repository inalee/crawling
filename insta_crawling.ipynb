```python
from selenium import webdriver  
from bs4 import BeautifulSoup as bs
import requests 
import re
import time as ti
driver = webdriver.Chrome('C:/Users/Ina Lee/Downloads/chromedriver_win32/chromedriver')
driver.implicitly_wait(3)           # 암묵적으로 웹 자원 로드를 위해 3초까지 기다려 준다.
emoji = re.compile('[\U00010000-\U0010ffff]',flags=re.UNICODE) #이모티콘 처리

keyword = "폴바셋"
link = "https://www.instagram.com/explore/tags/"+keyword
driver.get(link)
```

```python
from selenium.common.exceptions import WebDriverException,TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd

#emoji = re.compile('[\U00010000-\U0010ffff]',flags=re.UNICODE) #이모티콘 처리
#content = soup.select(".C4VMK > span")[0].text  ## [0] - 내용 / [1] - 댓글



xpath = '//*[@id="react-root"]/section/main/article/div[2]/div/div[1]/div[1]/a/div/div[2]'
driver.find_element_by_xpath(xpath).click()
driver.implicitly_wait(10) 
try:
    df = pd.DataFrame(columns=("content","time"))
    for num in range(10000):
        try:
            element = WebDriverWait(driver,10).until(
                # By.ID 는 ID로 검색, By.CSS_SELECTOR 는 CSS Selector 로 검색
                EC.presence_of_element_located((By.CLASS_NAME, "C4VMK"))     
            )
            
            html = driver.page_source
            soup = bs(html, 'html.parser') 
            content = soup.select(".k59kT")[0].text
            #hashtag = soup.select(".C4VMK > span")[1].text   ## [0] - 내용 / [1] - 댓글
            time = soup.select("time")[0].text
            print(content)
            print(time)
            print(num+1)
            df.loc[num] =[content,time]
            #driver.find_element_by_class_name("").click() #다음페이지
            driver.find_element_by_xpath("/html/body/div[3]/div/div[1]/div/div/a[2]").click()
            ti.sleep(1)
            num += 1    
            print("==========================================")

        except TimeoutException:
            print("타임아웃")
            driver.find_element_by_xpath("/html/body/div[3]/div/div[1]/div/div/a[2]").click()
            ti.sleep(1)
except WebDriverException:
    print("X")

```
